## Hypothesis testing for Stationarity

```python
from statsmodels.tsa.stattools import adfuller

results = adfuller(ts_passengers)

results
```

```python
## H0: Time series is non-stationary
## H1: Time Series is Stationary

### Accept Null Hypothesis if p >= 0.05
### Reject Null Hypothesis if p < 0.05

results = adfuller(res.resid)

results
```

```python
# ## H0: Time series is non-stationary
# ## H1: Time Series is Stationary
# ### Accept Null Hypothesis if p >= 0.05
# ### Reject Null Hypothesis if p < 0.05

from statsmodels.tsa.stattools import adfuller

def check_stationarity(timeseries):
  """
  Performs the Augmented Dickey-Fuller test to check for stationarity.

  Args:
    timeseries: A pandas Series representing the time series data.

  Returns:
    A tuple containing the test statistic, p-value, and other results.
    Prints whether the null hypothesis is accepted or rejected.
  """
  results = adfuller(timeseries)
  print('ADF Statistic: %f' % results[0])
  print('p-value: %f' % results[1])
  print('Critical Values:')
  for key, value in results[4].items():
    print('\t%s: %.3f' % (key, value))

  if results[1] >= 0.05:
    print("Accept Null Hypothesis: Time series is non-stationary")
  else:
    print("Reject Null Hypothesis: Time Series is Stationary")
  return results


check_stationarity(ts_passengers)
```

```python
model = ARIMA(ts_passengers[:100], order=(2,0,1))
model_fit = model.fit()

ts_test_pred = model_fit.get_forecast(steps=len(ts_passengers[100:])).predicted_mean

ts_passengers[:100].plot.line()
ts_passengers[100:].plot.line()
ts_test_pred.plot.line()
plt.legend(["Train", "Test", "Prediction"])
```
